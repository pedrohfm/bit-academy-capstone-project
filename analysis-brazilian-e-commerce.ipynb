# Brazilian e-commerce data manipulation

The goal is to match the portuguese product names to their english translations. For this we need to create a new column in the 'olist_products_dataset.csv' file.
We need to perform data manipulation operations outside of PowerBI which will facilitate the data analysis later on. 


```python
import pandas as pd

df_products = pd.read_csv("./analysis-brazilian-e-commerce-dataset/olist_products_dataset.csv")
df_translations = pd.read_csv("./analysis-brazilian-e-commerce-dataset/product_category_name_translation.csv")
```

## Checking data
Making sure the data is imported successfully.

```python
# Checking products dataset
df_products.info()
df_products
```

```python
# Checking translations dataset
df_translations.info()
df_translations
```

## Cleaning data

We don't need data on the dimensions, description or appearance of the products for our analysis. Let's get rid of those columns and then add the names in english. 

```python
# Getting rid of unecessary columns
df_products = df_products.drop(['product_name_lenght',
                'product_description_lenght',
                'product_photos_qty',
                'product_weight_g',
                'product_length_cm',
                'product_height_cm',
                'product_width_cm'],
                axis=1).copy()
df_products
```

Now we'll merge in the english names.

```python

products_dataset = pd.merge(left=df_products,
                            right=df_translations,
                            on='product_category_name',
                            how='left')
products_dataset
```

## It worked! Now we'll move it to PowerBI

We need to export a .csv file that will be used in PowerBi as a dataset for the product names in english and matching product IDs.

```python
products_dataset.to_csv("./analysis-brazilian-e-commerce-dataset/olist_products_dataset_translated.csv", index=False)
```
